```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# 1. Line plot for y = 2x + 3
x = np.linspace(-10, 10, 100)
y = 2 * x + 3
plt.figure()
plt.plot(x, y, label='y = 2x + 3', color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Line Plot of y = 2x + 3')
plt.legend()
plt.grid()
plt.show()

# 2. Sine and Cosine Wave
x = np.linspace(0, 2 * np.pi, 100)
sin_y = np.sin(x)
cos_y = np.cos(x)
plt.figure()
plt.plot(x, sin_y, label='sin(x)', color='red')
plt.plot(x, cos_y, label='cos(x)', color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sine and Cosine Waves')
plt.legend()
plt.grid()
plt.show()

# 3. Bar chart for city population
cities = ['Mumbai', 'Delhi', 'Bangalore', 'Hyderabad', 'Chennai']
population = [20, 18, 12, 10, 9]
plt.figure()
plt.bar(cities, population, color=['blue', 'green', 'red', 'purple', 'orange'])
plt.xlabel('Cities')
plt.ylabel('Population (millions)')
plt.title('Population of Cities')
plt.show()

# 4. Scatter plot
X = [2, 4, 6, 8, 10]
Y = [5, 15, 25, 35, 50]
colors = ['red', 'blue', 'green', 'purple', 'orange']
plt.figure()
plt.scatter(X, Y, color=colors)
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Scatter Plot')
plt.grid()
plt.show()

# 5. Histogram and Boxplot
random_data = np.random.normal(0, 1, 1000)
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].hist(random_data, bins=30, color='blue', edgecolor='black')
axes[0].set_title('Histogram')
axes[1].boxplot(random_data)
axes[1].set_title('Boxplot')
plt.show()

# 6. Three subplots (sine wave, bar chart, scatter plot)
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].plot(x, np.sin(x), color='blue')
axes[0].set_title('Sine Wave')
axes[1].bar(cities, population, color='green')
axes[1].set_title('Bar Chart')
axes[2].scatter(X, Y, color=colors)
axes[2].set_title('Scatter Plot')
plt.show()

# 7. Pie Chart
labels = ['Car', 'Bike', 'Bus', 'Train', 'Walking']
sizes = [40, 25, 20, 10, 5]
explode = (0.1, 0, 0, 0, 0)  # Exploding the Car slice
plt.figure()
plt.pie(sizes, labels=labels, autopct='%1.1f%%', explode=explode, colors=['red', 'blue', 'green', 'purple', 'orange'])
plt.title('Transport Mode Share')
plt.show()

# 8. Heatmap
random_matrix = np.random.rand(5, 5)
plt.figure()
plt.imshow(random_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.title('Random Heatmap')
plt.show()

# 9. Line chart for temperature data (CSV needed)
# Load the CSV file
df = pd.read_csv('temperatures.csv')

month = 'JAN'
plt.figure()
plt.plot(df['YEAR'], df[month], marker='o', linestyle='-', color='blue')
plt.xlabel('Year')
plt.ylabel(f'Temperature in {month} (°C)')
plt.title(f'Temperature Variation in {month} Over Years')
plt.grid()
plt.show()

# 10. Stacked Bar Chart for Sales
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
A = [10, 15, 20, 25]
B = [20, 25, 30, 35]
C = [30, 35, 40, 45]
plt.figure()
plt.bar(quarters, A, color='blue', label='Product A')
plt.bar(quarters, B, bottom=A, color='green', label='Product B')
plt.bar(quarters, C, bottom=np.array(A) + np.array(B), color='red', label='Product C')
plt.xlabel('Quarters')
plt.ylabel('Sales')
plt.title('Stacked Bar Chart of Product Sales')
plt.legend()
plt.show()
```
